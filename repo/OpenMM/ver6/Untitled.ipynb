```python
import numpy as np 
from scipy.integrate import odeint
import scipy
import matplotlib.pyplot as plt
```

```python
def func(y,t,conc,delta):
    I, D, U = y
    kf1 = 0.01 
    kf2 = 0.01
    kb1 = 0.05*conc
    kb2 = 0.2*delta
    
    dIdt = kf1*D + kf2*U - (kb1 + kb2)*I
    dDdt = kb2*I - kf1*D
    dUdt = kb1*I - kf2*U
    
    dydt = [dIdt, dDdt, dUdt]
    
    return dydt 

t = np.linspace(0,1000,10000)
yo = [1,0,0]

conc = np.linspace(0,1,10)
delta = np.linspace(0,0.5,10)

Ic = []
Dc = []
Uc = []

for i in np.arange(10):
    y = odeint(func,yo,t,args=(conc[i],delta[i]))
    Ic.append(y[:,0])
    Dc.append(y[:,1])
    Uc.append(y[:,2])
```

```python
plt.figure(figsize=(6,4),dpi=150)
for i in np.arange(10):
    plt.plot(t,Uc[i],label=str(i))
    plt.legend(loc=0)
```
