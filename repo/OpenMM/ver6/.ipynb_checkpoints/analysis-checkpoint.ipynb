```python
import numpy as np 
import yaml
import os 
import matplotlib.pyplot as plt

def csvread(name):
    dataraw = open(name+'.csv','r')
    
    # Empty dictionary for each column
    data_rmsd = {}
    counter = 0
    for line in dataraw:
        if line.strip():
            line = line.strip("\n' '")
            line = line.split(",")
            # Storing the data in the dictionary
            if counter == 0:
                data_rmsd['frame'] = []
                counter += 1
                for j in np.arange(len(line)):
                    data_rmsd[str(j)] = []
            for i in np.arange(len(line)):
                element = line[i]
                if i == 0:
                    data_rmsd['frame'].append(int(element))
                else:
                    data_rmsd[str(i)].append(float(element))

    return data_rmsd


def refined_rmsd_resting(name):
    data = csvread(name)
    
    collected = []
    for n in np.arange(len(data)-2):
        test = data[str(n+1)][-1]
        if test < 500:
            collected.append(n+1)
            
    counter = 0 
    
    for m in collected:
        if counter == 0:
            total = np.asarray(data[str(m)])
            counter += 1
        else:
            total = total + np.asarray(data[str(m)])
    
    avg = total/len(collected)
    
    msd = avg*avg
    
    return avg, msd, data['frame']

def refined_rmsd_activated(name):
    data = csvread(name)
    
    collected = []
    for n in np.arange(len(data)-2):
        test = data[str(n+1)][-1]
        if test < 100:
            collected.append(n+1)
            
    counter = 0 
    
    for m in collected:
        if counter == 0:
            total = np.asarray(data[str(m)])
            counter += 1
        else:
            total = total + np.asarray(data[str(m)])
    
    avg = total/len(collected)
    
    msd = avg*avg
    
    return avg, msd, data['frame']
```

```python
avg1, msd1, time1 = refined_rmsd('rmsdresting1')
avg2, msd2, time2 = refined_rmsd('rmsdresting2')
avg3, msd3, time3 = refined_rmsd('rmsdresting3')

avg1, msd1, time1 = refined_rmsd('rmsdresting1')
avg2, msd2, time2 = refined_rmsd('rmsdresting2')
avg3, msd3, time3 = refined_rmsd('rmsdresting3')
```

```python
plt.figure(figsize=(6,4),dpi=150)
plt.tick_params(direction='in',labelsize=10)
plt.title('RMSD')
plt.plot(time1,avg1,label='low activated cells')
plt.plot(time2,avg2,label='mid activated cells')
plt.plot(time3,avg3,label='high activated cells')
plt.xlabel('steps (x1000)')
plt.ylabel('distance (a.u.)')
plt.legend(loc=0)
```

```python
plt.figure(figsize=(6,4),dpi=150)
plt.tick_params(direction='in',labelsize=10)
plt.title('MSD')
plt.plot(time1,msd1,label='low activated cells')
plt.plot(time2,msd2,label='mid activated cells')
plt.plot(time3,msd3,label='high activated cells')
plt.xlabel('steps (x1000)')
plt.ylabel('distance (a.u.)')
plt.legend(loc=0)
```

```python
data1 = csvread('test_density1')
data2 = csvread('test_density2')
data3 = csvread('test_density3')

redata1 = {}
redata2 = {}
redata3 = {}
count = 1
for m in np.arange(len(data1['frame'])):
    if m%100 == 0:
        redata1[str(count)] = []
        redata2[str(count)] = []
        redata3[str(count)] = []
        for n in np.arange(len(data1)-2):
            element1 = data1[str(n+1)][m]
            element2 = data2[str(n+1)][m]
            element3 = data3[str(n+1)][m]
            redata1[str(count)].append(element1)
            redata2[str(count)].append(element2)
            redata3[str(count)].append(element3)
        count += 1 
```
