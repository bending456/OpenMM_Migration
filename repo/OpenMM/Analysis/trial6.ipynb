```python
os.chdir('/home/bending456/Ben-Code/Modeling-Codes/Codes/OpenMM_Migration/repo/OpenMM/trial7')
rawdata = open('outputDetail-Copy1.txt','r')
data = {}
counter = 0

for line in rawdata:
    if line.strip():
        #if counter == 0:
        #    counter += 1
        if counter == 0:
            counter += 1
            column_names = line.strip("\n' '")
            data_name = line.split(", ")
            for name in data_name:
                data[name] = []
        else:
            line = line.strip("\n' '")
            line = line.split(", ")
            for i in np.arange(len(data_name)):
                element = line[i]
                data[data_name[i]].append(element)
                
rawdata.close()

print(data)
```

```python
import matplotlib.pyplot as plt 
```

```python
num = [1, 50, 150]

ATPbyCell = [float(data['C_by_Cell'][0]),float(data['C_by_Cell'][2]),float(data['C_by_Cell'][4])]
ATPbyOrigin = [float(data['C_by_Origin'][0]),float(data['C_by_Origin'][2]),float(data['C_by_Origin'][4])]
ATPtotal = [float(data['C_total \n'][0]),float(data['C_total \n'][2]),float(data['C_total \n'][4])]

plt.figure(figsize=(6,4),dpi=150)
plt.tick_params(direction='in',labelsize=10)
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.plot(num,ATPbyCell,'ks:',lw=2,alpha=1,label='by Cell')
plt.plot(num,ATPbyOrigin,'bs:',lw=2,alpha=1,label='by Origin')
plt.plot(num,ATPtotal,'rs:',lw=2,alpha=1,label='Total')
plt.xlabel('Num of Activated Cells')
plt.ylabel('ATP Conc [a.u.]')
plt.legend(loc=0,fontsize=8)
plt.tight_layout()
```

```python
plt.figure(figsize=(6,4),dpi=150)
plt.tick_params(labelsize=12,direction='in')
langs = ['On', 'Off']
y_pos = np.arange(len(langs))
value = [float(data['C_by_Cell'][4]),float(data['C_by_Cell'][5])]
plt.bar(y_pos,value,0.5,alpha=1,color='gray',edgecolor='black')
plt.xticks(y_pos,langs)
plt.xlabel("State Variable On/Off",fontsize=12)
#plt.ylim([50,100])
plt.ylabel("ATP Conc [a.u.]",fontsize=12)
plt.tight_layout()
```
