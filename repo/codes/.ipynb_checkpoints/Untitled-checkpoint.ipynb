```python
import numpy as np 
from scipy.integrate import odeint
import scipy
import matplotlib.pyplot as plt
```

```python
def func(y,t,conc):
    kf = 0.001 
    kb = 0.001 
    
    dydt = kf*(1-y) - kb*conc*y
    
    return dydt 
```

```python
t = np.linspace(0,10000,10000)
conc = 0.5
y = odeint(func,1,t,args=(conc,))
if conc > 1:
    yt = (1-1/conc)*np.exp(-0.001*t-0.001*conc*t) + 0.001/(0.001+0.001*conc)
elif conc > 0.5 and conc < 1+1e-14:
    yt = 0.5*np.exp(-0.001*t-0.001*conc*t) + 0.001/(0.001+0.001*conc)
elif conc > 0.1 and conc < 0.5+1e-14:
    yt = 0.5*np.exp(-0.001*t-0.001*conc*t) + 0.001/(0.001+0.001*conc)
else:
    yt = conc*np.exp(-0.001*t-0.001*conc*t) + 0.001/(0.001+0.001*conc)
```

```python
plt.plot(t,y,label='odeint')
plt.plot(t,yt,label='exp.')
plt.legend(loc=0)
```
